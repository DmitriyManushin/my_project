В данном проекте я исследовал датасет о поддержаных автомобилях Toyota. Датасет содержит в себе следующую информацию: модель автомобиля, год  выпуска, стоимость, тип коробки передач, пробег, тип двигателя, стоимость налога на автомобиль в год, расход топлива и объем двигателя.

Бизнес-задача заключается в том, чтобы увеличить прибыль компании, которая занимается продажей поддержанных автомобилей Toyota.

Поднять продажи различных моделей автомобилей "Toyota" с пробегом для увеличения прибыли компании.



## Задачи

1. Проанализировать данные и выявить факторы (пробег, год выпуска, тип трансмиссии и т.д.), которые оказывают наибольшее влияние на цену автомобиля.

2. Выявить соотношение цены на автомобили разных моделей Toyota.

3. Построить "портрет" типичного потребителя, чтобы понять, какие модели автомобиля Toyota будут лучше всего продаваться, и напротив, на какие автомобили стоит снизить цены для увеличения продаж.

3. На основе анализа данных предложить рекомендации покупателям, какие модели и с какими техническими характеристиками являются наиболее выгодными для покупки.

Цели и задачи данного проекта направлены увеличение продаж автомобилей Toyota, а также на помощь людям, которые хотят купить поддержанный автомоболь Toyota в хорошем состоянии.

Анализ данных о продажах автомобилей Toyota с пробегом выявил две ключевые проблемы:

1. **Наличие старых автомобилей**. Эти автомобили требуют дополнительного времени и ресурсов на продажу, что может приводить к снижению рентабельности.
2. **Завышенные цены на некоторые модели**. Это может отпугивать потенциальных покупателей и снижать конкурентоспособность компании.

Таким образом, для увеличения прибыли от продажи автомобилей Toyota с пробегом можно предложить следующие варианты:
1. Проанализировать цены на автомобили, которые завышены в 1.5 - 2 по сравнению со средней стоимостью. Снизить цены до конкурентоспособного уровня, что увеличит объем продаж, а как следствие, выручку с продаж.
2. Определить факторы, которые могут приводить к завышенным ценам на некоторые модели, например, ошибки в оценке, недостаток информации о рыночной стоимости или особенности комплектации.
3. Разработать модель предсказания цены на автомобили Toyota с пробегом, учитывая факторы, выявленные в анализе. Это позволит более точно определять оптимальную цену продажи для каждого автомобиля.

В целом, для увеличения прибыли компании необходимо снизить цены на автомобили, которые не привлекают покупателей. Больше продаж означает больше денежных средств, поступающих от реализации автомобилей. Быстрая продажа автомобилей освобождает средства для приобретения новых автомобилей, что позволяет увеличить оборот денежных средств. В свою очередь, свободные денежные средства могут быть инвестированы в другие сферы бизнеса, например, в рекламу или расширение ассортимента.

# Общий вывод по проекту

В рамках проведенной работы было выполнено три ключевых этапа анализа данных об автомобилях Toyota с пробегом, а именно: предобработка данных, исследовательский анализ данных и графический анализ данных. Работа проводилась в онлайн-среде Google Colab на языке Python с использованием библиотек: NumPy, Pandas, Matplotlib и Seaborn.

**В первом разделе** был проведен анализ общей информации о данных. Исследованы основные параметры автомобилей: модель, год выпуска, цена, пробег, тип коробки передач, тип и объем двигателя, величина налога и расход топлива. Важной частью работы стало редактирование данных: проверка на наличие пропусков, корректировка ошибок и приведение данных к единому формату. Эти действия обеспечили удобство и простоту информации для дальшейней работы с данными.

**Во втором разделе** основной фокус был направлен на аналитическую обработку данных. Было проведено сравнение различных категорий автомобилей по их характеристикам, например, анализ цен в зависимости от года выпуска и пробега.
Также были выполнены вычисления средних значений (например, средний расход топлива, средний объем двигателя), что позволило более глубоко понять закономерности в данных.
В результате удалось выявить автомобили с отклонениями по цене (например, завышенные цены), а также определить группы автомобилей с высоким пробегом.
Также были даны рекомендации покупателям и представлены различных варианты для решения бизнес-задачи.

**В третьем разделе** работы была выполнена визуализация ключевых аспектов данных. Построены графики зависимости различных параметров, что дало возможность наглядно оценить динамику и распределение характеристик.
Дополнительно был сформирован портрет типичного потребителя, основанный на результатах анализа, что помогает лучше понять целевую аудиторию и использовать эти знания для разработки эффективных маркетинговых стратегий.
